# Preprocessing

```python
# Checking Missing Values
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum().sum()
```

```python
movies.duplicated().sum()
```

```python
import ast
def convert_genres(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"].apply(convert_genres)
```

```python
movies["genres"] = movies["genres"].apply(convert_genres)
movies.head()
```

```python
movies["keywords"] = movies["keywords"].apply(convert_genres)
movies.head()
```

```python
movies["cast"][0]
```

```python
# Taking Top 3 Casts
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if(counter!=3):
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
movies.head()
```

```python
movies["crew"][0]
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
        else:
            continue
    return L
```

```python
movies["crew"].apply(convert_crew)
```

```python
movies["crew"] = movies["crew"].apply(convert_crew)
movies.head()
```

```python
movies["overview"].apply(lambda x:x.split())
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
movies.head()
```

```python
# Removing Spaces
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["overview"] = movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
movies.head()
```

```python
movies["tags"] = movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
movies.head()
```

```python
new_df = movies[["id","title","tags"]]
new_df.head()
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:" ".join(x))
new_df.head()
```

```python
new_df["tags"][0]
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:str.lower(x))
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem_text(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem_text)
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
cv.get_feature_names_out()
```

# Cosine Similarity Computation

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

# Main Function for Recommendation

```python
def recommend(movie):
    # Fetching Index of the movie
    movie_index = new_df[new_df["title"] == movie].index[0]
    # Fetching the similarity scores
    distances = similarity[movie_index]
    # Using enumerate function to keep the index intact and then sorting on the basis of the second element of the enumerated list and then picking out 
    # 1 to 6 as 0 will be the movie itself
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Batman Begins")
```
